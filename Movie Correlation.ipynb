```python
# import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)  # Adjusts the configuration of plots


# Read in data
df = pd.read_csv(r'C:\Users\Mo\OneDrive - University of Arkansas\Portfolio\movies.csv')
```

```python
#Looking at the data

df.head()
```

```python
# checking for missing data

# Looking at % missing
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
    
# Alternatively, looking at number of missing rows in each column
for col in df.columns:
    print(df[col].isnull().value_counts(), "\n")
```

```python
#Dropping the rows with missing values
df = df.dropna()
```

```python
# Data types for columns

df.dtypes
```

```python
# Changing datatype of columns
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
#Create the correct year column from the released column
df['yearcorrect'] = df['released'].astype(str).str.split(',').str[-1].astype(str).str[:5]
```

```python
df.head()
```

```python
# Order by gross in descending order

df.sort_values(by=['gross'], inplace=False, ascending =False)
```

```python
#View the entire dataset

pd.set_option('display.max_rows', None)
```

```python
#Check and drop any duplicates

#df['company'].drop_duplicates().sort_values(ascending=False)

df.drop_duplicates()
```

```python
#Scatterplot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings') 

plt.show
```

```python
# Plot the budget vs Gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Looking at correlation

df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Feature')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#converting the categorical columns to categorical codes

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

        
df_numerized.head()
```

```python
#checking the correlation of all movie features including categorical

correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Feature')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
